```python
import pandas as pd
from pathlib import Path
```

```python
RAW_FILE = "../data/raw/gene_info"
SAVE_FILE = "../data/processed/gene_info_filtered.csv"
```

```python
import os

size = os.path.getsize(RAW_FILE) / (1024**3)
print(f"File size: {size:.2f} GB")
```

```python
chunksize = 500000

filtered_chunks = []

for chunk in pd.read_csv(
        RAW_FILE,
        sep="\t",
        chunksize=chunksize,
        dtype=str,
        low_memory=False):

    # filter human genes
    chunk = chunk[chunk["#tax_id"] == "9606"]

    # keep only useful columns
    chunk = chunk[[
        "GeneID",
        "Symbol",
        "description"
    ]]

    filtered_chunks.append(chunk)
```

```python
gene_df = pd.concat(filtered_chunks, ignore_index=True)

print("Total human genes:", len(gene_df))
gene_df.head()
```

```python
gene_df.to_csv(SAVE_FILE, index=False)

print("Saved:", SAVE_FILE)
```
